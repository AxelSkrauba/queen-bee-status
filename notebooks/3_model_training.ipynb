```python
import pycaret
pycaret.__version__
```

```python
import pandas as pd
from pycaret.classification import *
```

```python
# Evaluación con etiquetas originales
dataset = pd.read_csv("output_file_50mfcc_4000Hz.csv")
dataset = dataset.drop(columns=['file name'])

dataset_test = pd.read_csv("testoutput_file_50mfcc_4000Hz.csv")
dataset_test = dataset_test.drop(columns=['file name'])
```

```python
# PRIMERO
s = setup(data=dataset, test_data=dataset_test, index=False, target = 'label', session_id = 123,
          pca=True, pca_components=0.92)
```

```python
# Sin preprocesamiento, etiquetas reales  (PCA 0.92)
best = compare_models()
```

```python
# SEGUNDO
s = setup(data=dataset, test_data=dataset_test, index=False, target = 'label', session_id = 123,
          remove_outliers=True, fix_imbalance=True, normalize=True)
```

```python
# Con preprocesamiento, normalizado, remoción de outliers, SMOTE
best = compare_models()
```
